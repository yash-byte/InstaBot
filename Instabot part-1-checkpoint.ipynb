# Function for searching  the page

```python
def search_to_page(text):
    #Enabling Search box:
    #By clicking on it
    s=driver.find_element_by_class_name('Fifk5')
    s.click()
    time.sleep(4)
    s=driver.find_element_by_xpath('//input[contains(@class,"XTCLo ")]')
    #Clearing Input:
    s.clear()
    s.send_keys(text)
```

# Function for login to the page

```python
def login_to_page(text):
    s=driver.find_elements_by_class_name('Ap253')
    for i in s:
        if(i.text==text):
            i.click()
            break
```

# Function to like posts on a page

```python
def posts_like(n):
    s=driver.find_elements_by_class_name('_9AhH0')
    s[0].click()
    i=1
    while(i<=n):
        time.sleep(3)
        like_btn = driver.find_elements_by_xpath('//span[contains(@class,"fr66n")]/button/*[name()="svg"][@aria-label="Unlike"]')
        if(len(like_btn)==0):
            like=driver.find_element_by_xpath('//span[contains(@class,"fr66n")]/button')
            like.click()
        else:
            print('Post',i,'---------------->','Already Liked')
        i+=1
        if(i<=n):
            l=driver.find_element_by_xpath('//a[contains(@class,"coreSpriteRightPaginationArrow")]')
            l.click()
        time.sleep(2)
    close_btn = driver.find_element_by_xpath('//div[contains(@class,"BI4qX")]//button')
    close_btn.click()
```

# Function to unlike posts on a page

```python
def posts_unlike(n):
    s=driver.find_elements_by_class_name('_9AhH0')
    s[0].click()
    time.sleep(3)
    i=1
    while(i<=n):
        time.sleep(3)
        unlike_btn = driver.find_elements_by_xpath('//span[contains(@class,"fr66n")]/button/*[name()="svg"][@aria-label="Like"]')
        if(len(unlike_btn)==0):
            time.sleep(1)
            unlike=driver.find_element_by_xpath('//span[contains(@class,"fr66n")]/button')
            unlike.click()
        else:
            print('Post',i,'--------------->','Already UnLiked')
        i+=1
        if(i<=n):
            l=driver.find_element_by_xpath('//a[contains(@class,"coreSpriteRightPaginationArrow")]')
            l.click()
        time.sleep(2)
    close_btn = driver.find_element_by_xpath('//div[contains(@class,"BI4qX")]//button')
    close_btn.click()
```

# Function to print name of  First N followers of a page

```python
from selenium.common.exceptions import StaleElementReferenceException
def follower(n):
    driver.find_element_by_xpath('//ul[contains(@class,"k9GMp")]/li[2]/a').click()
    time.sleep(3)
    while True:
        try:
            followers=[]
            count=0
            while(True):
                elements=driver.find_elements_by_xpath('//a[contains(@class, "notranslate")]')
                if(len(elements)<n):
                    driver.execute_script('arguments[0].scrollIntoView(0, 100);', driver.find_element_by_xpath('//ul[contains(@class, "jSC57")]'))
                    time.sleep(1)    
                if(count<len(elements)):
                    i=elements[count]
                    count+=1
                    followers.append(i.get_attribute('innerHTML'))                          
                if(len(followers)>=n):
                    break
            break    
        except StaleElementReferenceException:
            continue
    driver.back()        
    return followers
```

# Function to get Following list of a page

```python
def following(n):
    s=driver.find_element_by_xpath('//ul[contains(@class,"k9GMp")]/li[3]/a')
    s.click()
    time.sleep(3)
    while True:
        try:
            following=[]
            count=0
            while(True):
                elements=driver.find_elements_by_xpath('//a[contains(@class, "notranslate")]')
                if(len(elements)<n):
                    driver.execute_script('arguments[0].scrollIntoView(0, 100);', driver.find_element_by_xpath('//ul[contains(@class, "jSC57")]'))
                    time.sleep(1)
                if(count<len(elements)):
                    i=elements[count]
                    count+=1
                    following.append(i.text)
                if(len(following)>=n):
                    break
            break    
        except StaleElementReferenceException:
            continue
    driver.back()        
    return following
```
